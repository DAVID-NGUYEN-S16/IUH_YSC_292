```python
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
from selenium.webdriver.chrome.options import Options
import pandas as pd
import numpy as np
import random
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer
%matplotlib inline
```

```python
def handlePopups():
    option = Options()
    option.add_argument("disable-notifications")
    return option

driver = webdriver.Chrome(executable_path="/Users/nguye/YSC/NCKH/YSC/chromedriver.exe", options=handlePopups())
driver.get("https://www.google.com/maps/")
```

```python
def searchplace(address):
    # Tìm class input truy cập đến nó
    place = driver.find_element_by_class_name("tactile-searchbox-input")
    # Gửi địa chỉ thực vào để truy vấn 
    place.send_keys(address)
    # tìm đến thanh tìm kiếm 
    # place.find_element_by_tag_name('html').send_key(Keys.ENTER)
    sub = driver.find_element_by_xpath('/html/body/div[3]/div[9]/div[3]/div[1]/div[1]/div[1]/div[2]/div[1]/button')
    # Nhấn click
    sub.click()
    # place.clear()
```

```python
def get_url(url):
    kq = (re.findall(r'[0-9]{1,}.[0-9]{1,}!4d[0-9]{1,}.[0-9]{1,}', url))
    # print(kq)
    if len(kq) == 0:
        kq = re.findall(r'[0-9]{1,}.[0-9]{1,},[0-9]{1,}.[0-9]{1,}', url)
        latitude, longtitude = kq[0].split(",")
    else:
        latitude, longtitude = kq[0].split("!4d")

    latitude = float(latitude)
    longtitude = float(longtitude)
    driver.find_element_by_id('searchboxinput').clear()
    # print(latitude,longtitude)
    return latitude, longtitude
```

```python
file_address = pd.read_excel('ADDRESS.xlsx')
# file_address
```

```python
import random
id = 0 
store_lat = []
store_log = []
for x in file_address['Địa chỉ']:
    # time.sleep(0.5) 
    searchplace(x)
    sleep(random.uniform(1.1, 2))
    st = get_url(driver.current_url)
    print(st)
    store_lat.append(st[0])
    store_log.append(st[1])
```

```python
# file_address.insert()
file_address['Latitude'] = store_lat
file_address['Logtitude'] = store_log
file_address
```
